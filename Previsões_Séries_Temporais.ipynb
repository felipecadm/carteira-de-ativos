```python
from prophet import Prophet
from prophet.plot import plot_plotly
import pandas as pd
import numpy as np
from sklearn.metrics import mean_absolute_error
```

```python
edp = pd.read_csv('acoes.csv', usecols=['Date', 'EDP']).rename(columns={'Date': 'ds', 'EDP': 'y'})
itsa = pd.read_csv('acoes.csv', usecols=['Date', 'ITSA']).rename(columns={'Date': 'ds', 'ITSA': 'y'})
petr = pd.read_csv('acoes.csv', usecols=['Date', 'PETR']).rename(columns={'Date': 'ds', 'PETR': 'y'})
ibov = pd.read_csv('acoes.csv', usecols=['Date', 'IBOV']).rename(columns={'Date': 'ds', 'IBOV': 'y'})
#conforme padrão exigido pela biblioteca prophet
```

```python
#Criação do modelo - por ativo - e treinamento
modelo_edp = Prophet().fit(edp)
modelo_itsa = Prophet().fit(itsa)
modelo_petr = Prophet().fit(petr)
modelo_ibov = Prophet().fit(ibov)
```

```python
#Construção de novo dataframe e previsão para 90 dias da data atual
future_edp = modelo_edp.make_future_dataframe(periods=90)
future_itsa = modelo_itsa.make_future_dataframe(periods=90)
future_petr = modelo_petr.make_future_dataframe(periods=90)
future_ibov = modelo_ibov.make_future_dataframe(periods=90)

previsoes_edp = modelo_edp.predict(future_edp)
previsoes_itsa = modelo_itsa.predict(future_itsa)
previsoes_petr = modelo_petr.predict(future_petr)
previsoes_ibov = modelo_ibov.predict(future_ibov)
```

##**Gráficos - por ativo - de Modelo vs Resultados & linhas de tendência**

###**EDP**

```python
plot_plotly(modelo_edp, previsoes_edp, xlabel='', ylabel='')
```

```python
#Erro Médio Absoluto do Modelo para previsões
mean_absolute_error(previsoes_edp['yhat'].head(edp.shape[0]), edp['y'])
```

###**ITSA**

```python
plot_plotly(modelo_itsa, previsoes_itsa, xlabel='', ylabel='')
```

```python
mean_absolute_error(previsoes_itsa['yhat'].head(itsa.shape[0]), itsa['y'])
```

###**PETR**

```python
plot_plotly(modelo_petr, previsoes_petr, xlabel='', ylabel='')
```

```python
mean_absolute_error(previsoes_petr['yhat'].head(petr.shape[0]), petr['y'])
```

###**IBOV**

```python
plot_plotly(modelo_ibov, previsoes_ibov, xlabel='', ylabel='')
```
